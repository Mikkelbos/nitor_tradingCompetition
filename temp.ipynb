```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('train.csv')

# Convert delivery_start to datetime
df['delivery_start'] = pd.to_datetime(df['delivery_start'])

# Sort by delivery_start to ensure the plot makes sense line-wise
df = df.sort_values('delivery_start')

# Plotting
plt.figure(figsize=(15, 7))
plt.plot(df['delivery_start'], df['target'], label='Target', alpha=0.7)
plt.xlabel('Delivery Start')
plt.ylabel('Target')
plt.title('Target vs Delivery Start')
plt.legend()
plt.grid(True)
plt.show()
```
